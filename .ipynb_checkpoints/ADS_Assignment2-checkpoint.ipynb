```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df=pd.read_csv("training.csv")
test_df=pd.read_csv("testing.csv")
```

In the training dataset, there are 32 columns and there is a scope of column transformation. WeekStatus and Days_of_week can be transformed to another columns and there is no null data in the dataset. Also, the column Rv1 and Rv2 are similar, we can drop either Rv1 or Rv2. Also, date column is not required because NSM column already got interpreted from this. 

```python
train_df=train_df.drop(['rv1','date'],axis=1)
test_df=test_df.drop(['rv1','date'],axis=1)
```

Transformation of WeekStatus and Days_of_week columns 

```python
weekstatus_train=pd.get_dummies(train_df['WeekStatus'])
days_of_week_train=pd.get_dummies(train_df['Day_of_week'])

weekstatus_test=pd.get_dummies(test_df['WeekStatus'])
days_of_week_test=pd.get_dummies(test_df['Day_of_week'])

#Concatenation of the above variable to the main dataframe

train_df=pd.concat((train_df,weekstatus_train),axis=1)
train_df=pd.concat((train_df,days_of_week_train),axis=1)


test_df=pd.concat((test_df,weekstatus_test),axis=1)
test_df=pd.concat((test_df,days_of_week_test),axis=1)


#drop the WeekStatus and Day_of_week column

train_df=train_df.drop(['WeekStatus','Day_of_week'],axis=1)
test_df=test_df.drop(['WeekStatus','Day_of_week'],axis=1)
```

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

lr= LinearRegression()

#Calculate the training dataset's metrices

X=train_df.drop(['Appliances'],axis=1)
y=train_df['Appliances']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

lr.fit(X_train,y_train)
prediction=lr.predict(X_test)

rms = sqrt(mean_squared_error(y_test, prediction))
r2= r2_score(y_test, prediction)
mae=mean_absolute_error(y_test,prediction)
mape=np.mean(np.abs((y_test - prediction) / y_test)) * 100

#Calculate the testing dataset's metrices

A=test_df.drop(['Appliances'],axis=1)
b=test_df['Appliances']

A_train, A_test, b_train, b_test = train_test_split(A, b, test_size=0.3)

lr.fit(A_train,b_train)
prediction_test_lr=lr.predict(A_test)

rms_test_lr = sqrt(mean_squared_error(b_test, prediction_test_lr))
r2_test_lr= r2_score(b_test, prediction_test_lr)
mae_test_lr=mean_absolute_error(b_test,prediction_test_lr)
mape_test_lr=np.mean(np.abs((b_test - prediction_test_lr) / b_test)) * 100

#printing the training and testing metrices
print('RMS of Training Data in Linear Regression:',rms)
print('RMS of Testing Data in Linear Regression:',rms_test_lr)
print('R Squared of Training Data in Linear Regression:',r2)
print('R Squared of Testing Data in Linear Regression:',r2_test_lr)
print('MAE of Training Data in Linear Regression:',mae)
print('MAE of Testing Data in Linear Regression:',mae_test_lr)
print('MAPE of Training Data in Linear Regression:',mape)
print('MAPE of Testing Data in Linear Regression:',mape_test_lr)

```

#### Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=500)

#for training dataset

rf.fit(X_train, y_train)

predict_train_rf=rf.predict(X_test)

rms_train_rf = sqrt(mean_squared_error(y_test, predict_train_rf))
r2_train_rf= r2_score(y_test, predict_train_rf)
mae_train_rf=mean_absolute_error(y_test,predict_train_rf)
mape_train_rf=np.mean(np.abs((y_test - predict_train_rf) / y_test)) * 100

#for testing dataset

rf.fit(A_train, b_train)

predict_test_rf=rf.predict(A_test)

rms_test_rf = sqrt(mean_squared_error(b_test, predict_test_rf))
r2_test_rf= r2_score(b_test, predict_test_rf)
mae_test_rf=mean_absolute_error(b_test,predict_test_rf)
mape_test_rf=np.mean(np.abs((b_test - predict_test_rf) / b_test)) * 100

#printing the training and testing metrices
print('RMS of Training Data in Random Forest :',rms_train_rf)
print('RMS of Testing Data in Random Forest :',rms_test_rf)
print('R Squared of Training Data in Random Forest :',r2_train_rf)
print('R Squared of Testing Data in Random Forest :',r2_test_rf)
print('MAE of Training Data in Random Forest :',mae_train_rf)
print('MAE of Testing Data in Random Forest :',mae_test_rf)
print('MAPE of Training Data in Random Forest :',mape_train_rf)
print('MAPE of Testing Data in Random Forest :',mape_test_rf)
```

#### Neural Network 

```python
#import the Multi-Layer Perceptron Classifier model
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(37,37,37))

# for training dataset
mlp.fit(X_train,y_train)
prediction_train_nn = mlp.predict(X_test)

rms_train_nn = sqrt(mean_squared_error(y_test, prediction_train_nn))
r2_train_nn= r2_score(y_test, prediction_train_nn)
mae_train_nn=mean_absolute_error(y_test,prediction_train_nn)
mape_train_nn=np.mean(np.abs((y_test - prediction_train_nn) / y_test)) * 100

# for testing dataset
mlp.fit(A_train,b_train)
prediction_test_nn = mlp.predict(A_test)

rms_test_nn = sqrt(mean_squared_error(b_test, prediction_test_nn))
r2_test_nn= r2_score(b_test, prediction_test_nn)
mae_test_nn=mean_absolute_error(b_test,prediction_test_nn)
mape_test_nn=np.mean(np.abs((b_test - prediction_test_nn) / b_test)) * 100

#printing the training and testing metrices
print('RMS of Training Data in Neural Network:',rms_train_nn)
print('RMS of Testing Data in Neural Network:',rms_test_nn)
print('R Squared of Training Data in Neural Network:',r2_train_nn)
print('R Squared of Testing Data in Neural Network:',r2_test_nn)
print('MAE of Training Data in Neural Network:',mae_train_nn)
print('MAE of Testing Data in Neural Network:',mae_test_nn)
print('MAPE of Training Data in Neural Network:',mape_train_nn)
print('MAPE of Testing Data in Neural Network:',mape_test_nn)
```

#### Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import (LinearRegression, Ridge, Lasso, RandomizedLasso)


# Define dictionary to store our rankings
ranks = {}
#colnames = train_df.columns
colnames= ['lights', 'T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4',
       'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9',
       'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility',
       'Tdewpoint', 'rv2', 'NSM', 'Weekday', 'Weekend', 'Friday', 'Monday',
       'Saturday', 'Sunday', 'Thursday', 'Tuesday','Wednesday']



# Create our function which stores the feature rankings to the ranks dictionary
def ranking(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x,2), ranks)
    return dict(zip(names, ranks))

rfe = RFE(lr, n_features_to_select=1, verbose =3 )
rfe.fit(X,y)
ranks["RFE"] = ranking(list(map(float, rfe.ranking_)), colnames, order=-1)

# Using Linear Regression
lr = LinearRegression(normalize=True)
lr.fit(X,y)
ranks["LinReg"] = ranking(np.abs(lr.coef_), colnames)

# Using Ridge 
ridge = Ridge(alpha = 7)
ridge.fit(X,y)
ranks['Ridge'] = ranking(np.abs(ridge.coef_), colnames)

# Using Lasso
lasso = Lasso(alpha=.05)
lasso.fit(X, y)
ranks["Lasso"] = ranking(np.abs(lasso.coef_), colnames)

#Using Random Forest Regressor
rf = RandomForestRegressor(n_jobs=-1, n_estimators=50, verbose=3)
rf.fit(X,y)
ranks["RF"] = ranking(rf.feature_importances_, colnames);

#creating the feature ranking matrix
# Create empty dictionary to store the mean value calculated from all the scores
r = {}
for name in colnames:
    r[name] = round(np.mean([ranks[method][name] for method in ranks.keys()]), 2)

methods = sorted(ranks.keys())
ranks["Mean"] = r
methods.append("Mean")    
```

```python
print("\t%s" % "\t".join(methods))
for name in colnames:
    print("%s\t%s" % (name, "\t".join(map(str, 
                         [ranks[method][name] for method in methods]))))
```

```python
# Put the mean scores into a Pandas dataframe
meanplot = pd.DataFrame(list(r.items()), columns= ['Feature','Mean Ranking'])

# Sort the dataframe
meanplot = meanplot.sort_values('Mean Ranking', ascending=False)

# Let's plot the ranking of the features
sns.factorplot(x="Mean Ranking", y="Feature", data = meanplot, kind="bar", 
               size=14, aspect=1.9, palette='coolwarm')
```
